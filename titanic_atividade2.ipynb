```python
from seaborn import load_dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier


#tratando dados
df = load_dataset('titanic')
df = df.drop(columns=['class', 'deck'])

df.dropna(inplace=True)
df = pd.get_dummies(df, columns=['sex', 'embarked', 'who', 'embark_town', 'alive'])
df.head()

#Dados tratados
```

```python
#separando
columns = len(df.columns)

y_pclass = df['pclass']
x_pclass=df.drop('pclass', axis=1)

y_survived = df['survived']
x_survived = df.drop('survived', axis=1)
```

```python
#dividindo os datasets

k=5

x_pclass_train, x_pclass_test, y_pclass_train, y_pclass_test  = train_test_split(x_pclass, y_pclass, test_size=0.2, random_state=None, stratify=y_pclass)
x_survived_train, x_survived_test, y_survived_train, y_survived_test  = train_test_split(x_survived, y_survived, test_size=0.2, random_state=None, stratify=y_survived)
```

```python
# Treinamendo da Árvore de Decisão(AD)
model_pclass_AD = tree.DecisionTreeClassifier(criterion="entropy")
model_pclass_AD.fit(x_pclass_train, y_pclass_train)

model_survived_AD = tree.DecisionTreeClassifier(criterion="entropy")
model_survived_AD.fit(x_survived_train, y_survived_train)

# Treinamento KNN

model_pclass_KNN = KNeighborsClassifier(n_neighbors=k, metric='euclidean', algorithm='brute')
model_pclass_KNN = model_pclass_KNN.fit(x_pclass_train, y_pclass_train)

model_survived_KNN = KNeighborsClassifier(n_neighbors=k, metric='euclidean', algorithm='brute')
model_survived_KNN = model_survived_KNN.fit(x_survived_train, y_survived_train)
```

```python
# Predição e Resultados em Árvore de Decisão

# Exemplo avaliando em pclass
result_pclass_AD = model_pclass_AD.predict(x_pclass_test)
acc_pclass = metrics.accuracy_score(result_pclass_AD, y_pclass_test)
print("Acurácia no teste de para a classe 'pclass' usando arvore de decisão: {:.2f}%".format(acc_pclass * 100))

# Exemplo avaliando em survived

result_survived_AD = model_survived_AD.predict(x_survived_test)
acc_survived = metrics.accuracy_score(result_survived_AD, y_survived_test)
print("Acurácia no teste de para a classe 'survived' usando arvore de decisão: {:.2f}%".format(acc_survived * 100))


# Predição e Resultados em KNN

# Exemplo avaliando em pclass
result_pclass_KNN = model_pclass_KNN.predict(x_pclass_test)
acc_pclass = metrics.accuracy_score(result_pclass_KNN, y_pclass_test)
print("Acurácia no teste de para a classe 'pclass' usando KNN: {:.2f}%".format(acc_pclass * 100))

# Exemplo avaliando em survived

result_survived_KNN = model_survived_KNN.predict(x_survived_test)
acc_survived = metrics.accuracy_score(result_survived_KNN , y_survived_test)
print("Acurácia no teste de para a classe 'survived' usando KNN: {:.2f}%".format(acc_survived * 100))

```
