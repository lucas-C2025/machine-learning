```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler

url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/transfusion.data"

# Carregar base de dados
df = pd.read_csv(url, header=None)

#criando escala

scaler = MinMaxScaler()

df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

columns = len(df.columns)
y_0 = df[0] # extrai a primeira coluna, que é o label
X_0 = df.loc[:,1:columns-1]


y_2 = df[0] # extrai a primeira coluna, que é o label
X_2 = df.loc[:,1:columns-1]

df.head()

```

```python
#dividindo os datasets

k=5

X_0_train, X_0_test, y_0_train, y_0_test = train_test_split(X_0, y_0, test_size=0.2, random_state=None, stratify=y_0)
X_2_train, X_2_test, y_2_train, y_2_test = train_test_split(X_2, y_2, test_size=0.2, random_state=None, stratify=y_2)
```

```python
# Treinamendo da Árvore de Decisão(AD)
model_X_0_AD = tree.DecisionTreeClassifier(criterion="entropy")
model_X_0_AD.fit(X_0_train, y_0_train)

model_X_2_AD = tree.DecisionTreeClassifier(criterion="entropy")
model_X_2_AD.fit(X_2_train, y_2_train)

# Treinamento KNN

model_X_0_KNN = KNeighborsClassifier(n_neighbors=k, metric='euclidean', algorithm='brute')
model_X_0_KNN = model_X_0_KNN.fit(X_0_train, y_0_train)

model_X_2_KNN = KNeighborsClassifier(n_neighbors=k, metric='euclidean', algorithm='brute')
model_X_2_KNN = model_X_2_KNN.fit(X_2_train, y_2_train)
```

```python
# Predição e Resultados em Árvore de Decisão

# Exemplo avaliando em X_0
result_X_0_AD = model_X_0_AD.predict(X_0_test)
acc_X_0_AD = metrics.accuracy_score(y_0_test, result_X_0_AD) # Correção aqui
print("Acurácia no teste de para a classe 'X_0' usando arvore de decisão: {:.2f}%".format(acc_X_0_AD * 100))

# Exemplo avaliando em X_2

result_X_2_AD = model_X_2_AD.predict(X_2_test)
acc_X_2_AD = metrics.accuracy_score(y_2_test, result_X_2_AD) # Correção aqui
print("Acurácia no teste de para a classe 'X_2' usando arvore de decisão: {:.2f}%".format(acc_X_2_AD * 100))


# Predição e Resultados em KNN

# Exemplo avaliando em X_0
result_X_0_KNN = model_X_0_KNN.predict(X_0_test)
acc_X_0_KNN = metrics.accuracy_score(y_0_test, result_X_0_KNN) # Correção aqui
print("Acurácia no teste de para a classe 'X_0' usando KNN: {:.2f}%".format(acc_X_0_KNN * 100))

# Exemplo avaliando em X_2

result_X_2_KNN = model_X_2_KNN.predict(X_2_test)
acc_X_2_KNN = metrics.accuracy_score(y_2_test, result_X_2_KNN) # Correção aqui
print("Acurácia no teste de para a classe 'X_2' usando KNN: {:.2f}%".format(acc_X_2_KNN * 100))
```
